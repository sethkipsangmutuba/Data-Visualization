##1.2 Introducing Pandas Objects

At the very basic level, Pandas objects can be thought of as enhanced versions of NumPy structured arrays in which the rows and columns are identified with labels rather than simple integer indices. Pandas provides a host of useful tools, methods, and functionality on top of the basic data structures, but nearly everything that follows will require an understanding of what these structures are. Thus, before going further, let’s introduce three fundamental Pandas data structures: the Series, DataFrame, and Index.

We will start with the standard NumPy and Pandas imports.

---

##1.2.1 The Pandas Series Object

A Pandas Series is a one-dimensional array of indexed data. It can be created from a list or array.

As seen, the Series wraps both a sequence of values and a sequence of indices, which can be accessed with the `values` and `index` attributes. The values are simply a familiar NumPy array.

The index is an array-like object of type `pd.Index`.

Like with a NumPy array, data can be accessed by the associated index via Python square-bracket notation.

The Pandas Series is more general and flexible than the one-dimensional NumPy array that it emulates.

---

### Series as generalized NumPy array

The key difference is the presence of the index: while the NumPy array has an implicitly defined integer index, the Pandas Series has an explicitly defined index associated with the values.

This explicit index definition gives the Series object additional capabilities. The index need not be an integer, but can consist of values of any desired type, such as strings.

Item access works as expected, and indices can also be noncontiguous or nonsequential.

---

### Series as specialized dictionary

A Pandas Series can be viewed as a specialization of a Python dictionary. A dictionary maps arbitrary keys to values, while a Series maps typed keys to typed values. This typing makes it more efficient than Python dictionaries for certain operations.

This analogy becomes clearer by constructing a Series directly from a dictionary. By default, the index is drawn from the sorted keys, and dictionary-style item access can be performed.

Unlike a dictionary, the Series also supports array-style operations such as slicing.

---

###Constructing Series objects

A Pandas Series can be constructed as:

`pd.Series(data, index=index)`

where `index` is optional, and `data` can take various forms.

Data can be a list or NumPy array, in which case the index defaults to an integer sequence.

Data can be a scalar, which is repeated to fill the specified index.

Data can be a dictionary, in which case the index defaults to the sorted dictionary keys.

In each case, the index can be explicitly set if a different result is preferred.

In such cases, the Series is populated only with the explicitly identified keys.

##1.2.2 The Pandas DataFrame Object

The next fundamental structure in Pandas is the DataFrame. Like the Series object, the DataFrame can be thought of either as a generalization of a NumPy array, or as a specialization of a Python dictionary.

---

### DataFrame as a generalized NumPy array

If a Series is an analog of a one-dimensional array with flexible indices, a DataFrame is an analog of a two-dimensional array with both flexible row indices and flexible column names. Just as a two-dimensional array can be seen as an ordered sequence of aligned one-dimensional columns, a DataFrame can be viewed as a sequence of aligned Series objects that share the same index.

To demonstrate this, we construct a Series listing the area of each state and combine it with the population Series to form a two-dimensional object using a dictionary.

Like the Series object, the DataFrame has an `index` attribute that provides access to row labels. It also has a `columns` attribute, which is an Index object holding the column labels.

Thus, the DataFrame generalizes a two-dimensional NumPy array by allowing both rows and columns to have labeled indices.

---

### DataFrame as specialized dictionary

A DataFrame can also be viewed as a specialization of a dictionary. Instead of mapping a key to a single value, it maps a column name to a Series of column data. Accessing a column returns the corresponding Series.

This leads to an important distinction: in a NumPy array, accessing data by index typically returns rows, whereas in a DataFrame, accessing by a column name returns columns. For this reason, it is often more intuitive to think of DataFrames as generalized dictionaries.

---

### Constructing DataFrame objects

A Pandas DataFrame can be constructed in several ways:

- From a single Series object, resulting in a one-column DataFrame  
- From a list of dictionaries, where each dictionary represents a row; missing keys are filled with NaN values  
- From a dictionary of Series objects, aligning data by index  
- From a two-dimensional NumPy array, with optional specification of column and row labels  
- From a NumPy structured array, where fields map naturally to DataFrame columns

##1.2.3 The Pandas Index Object

Both the Series and DataFrame objects contain an explicit index that lets you reference and modify data. This Index object can be thought of either as an immutable array or as an ordered set (technically a multiset, since it may contain repeated values). These views influence the operations available on Index objects.

---

###Index as immutable array

The Index object behaves in many ways like an array. Standard Python indexing can be used to retrieve values or slices.

Index objects also expose familiar attributes such as size, shape, dimensionality, and data type.

A key difference from NumPy arrays is immutability: Index objects cannot be modified through assignment. This property makes them safer to share across multiple DataFrames and arrays without unintended side effects.

---

###Index as ordered set

Pandas objects support operations such as joins that rely on set-like behavior. The Index object follows conventions similar to Python’s set structure, enabling operations like unions, intersections, and differences.

These set operations can also be accessed through object methods such as `intersection`, providing flexible ways to combine and compare indices.

##1.3 Data Indexing and Selection

Previously, methods for accessing and modifying values in NumPy arrays were introduced, including direct indexing, slicing, masking, fancy indexing, and combinations of these approaches.

Similar mechanisms exist for Pandas Series and DataFrame objects. If you are familiar with NumPy-style indexing, the corresponding Pandas operations will feel familiar, though there are important differences to keep in mind.

We begin with the one-dimensional Series structure, then extend the discussion to the two-dimensional DataFrame structure.

##1.3.1 Data Selection in Series

A Series object behaves both like a one-dimensional NumPy array and like a Python dictionary. These two mental models help explain how indexing and selection work in practice.

---

### Series as dictionary

A Series provides a mapping between keys (index labels) and values.

It supports dictionary-like access to values using keys. It also allows inspection of keys and values using familiar dictionary-style methods.

A Series can be modified dynamically by assigning values to new keys, similar to extending a dictionary. This mutability is managed internally by Pandas.

---

### Series as one-dimensional array

A Series also supports array-like operations such as slicing, masking, and fancy indexing, similar to NumPy arrays.

Slicing behavior depends on whether explicit or implicit indexing is used. When slicing with labels, the final element is included, while integer-based slicing follows standard Python conventions where the endpoint is excluded.

---

### Indexers: loc, iloc, and ix

To reduce ambiguity, Pandas provides explicit indexing operators.

- `loc` selects data using explicit index labels.  
- `iloc` selects data using integer-based positional indexing.  
- `ix` was introduced as a hybrid of both approaches, but is now largely deprecated in favor of clearer use of `loc` and `iloc`.

Using explicit indexers improves readability and reduces errors, especially when working with integer-indexed Series.

##1.3.2 Data Selection in DataFrame

A DataFrame can be viewed either as a two-dimensional structured array or as a dictionary of aligned Series sharing the same index.

---

## DataFrame as a dictionary

A DataFrame is often interpreted as a collection of Series objects (columns). Each column can be accessed using dictionary-style indexing or attribute-style access when the column name is valid.

Column access returns the corresponding Series, and arithmetic between Series allows creation of new columns, such as computing derived values like density from population and area.

Attribute-style access is a convenient shortcut but is not always safe, especially when column names conflict with built-in methods or are not valid identifiers.

---

## DataFrame as a two-dimensional array

A DataFrame can also be viewed as a matrix-like structure. The underlying data can be accessed as a NumPy array, and it supports operations like transposition.

However, unlike a pure NumPy array, direct indexing behaves differently: single indexing typically refers to columns rather than rows, due to label-based structure.

To handle both row and column selection consistently, Pandas provides explicit indexers:

- `iloc`: position-based (integer indexing like NumPy)  
- `loc`: label-based (index and column names)  
- `ix`: hybrid approach (older and discouraged due to ambiguity)  

These indexers support slicing, masking, and combined row/column selection.

---

## Selection and modification

DataFrames support boolean filtering and mixed indexing, allowing conditions to select specific rows and columns. Values can also be updated directly using `loc` or `iloc`.

---

## Additional conventions

- Column selection uses indexing syntax  
- Row slicing uses standard slice notation  
- Boolean masks are applied row-wise  

These conventions make DataFrame operations flexible but require careful attention to avoid confusion between row and column selection.

##1.4 Operating on Data in Pandas

Pandas builds on NumPy’s element-wise computation model and fully supports universal functions (ufuncs) for fast numerical operations.

For single-input operations (such as negation or mathematical functions), Pandas preserves index and column labels in the output, ensuring that structure is maintained alongside computation.

For binary operations (such as addition or multiplication), Pandas automatically aligns data by index before performing the operation. This means values are matched based on labels rather than position, which reduces errors when combining datasets from different sources.

This alignment behavior makes data integration more robust, especially when working with heterogeneous or partially overlapping datasets.

Pandas also supports operations between Series and DataFrame objects, extending these element-wise and alignment-based behaviors to both one-dimensional and two-dimensional structures.

##1.4.1 Ufuncs: Index Preservation

Pandas is fully compatible with NumPy’s universal functions (ufuncs). This means any ufunc applied to a Series or DataFrame produces a Pandas object with the original index and structure preserved.

When applied to a Series, the operation is performed element-wise while retaining the index labels. When applied to a DataFrame, the operation is applied element-wise across the entire table, preserving both row and column labels.

Even more complex transformations, such as trigonometric functions or scaling operations, behave in the same way. The output remains aligned with the original structure, ensuring interpretability of results without losing contextual labels.

SERIES + UFUNC (INDEX PRESERVATION)

```python
sepal_length = pd.Series(
    df["sepal length (cm)"].values,
    index=df.index
)

print("Original Series (first 5):\n", sepal_length.head())

# Apply NumPy ufunc
log_series = np.log(sepal_length)

print("\nAfter np.log (index preserved):\n", log_series.head())
```

MULTIPLE UFUNCS ON SERIES

```python
scaled_series = np.sqrt(sepal_length) + np.sin(sepal_length)

print("\nComposite ufunc result:\n", scaled_series.head())
```

 DATAFRAME-WIDE UFUNC APPLICATION

```python
df_numeric = df.iloc[:, :4]

log_df = np.log(df_numeric)

print("\nDataFrame after np.log (structure preserved):\n", log_df.head())
```

TRIGONOMETRIC UFUNCS ON DATAFRAME

```python
sin_df = np.sin(df_numeric)

print("\nnp.sin applied to DataFrame:\n", sin_df.head())
```

SCALING TRANSFORMATION (scalar broadcasting + ufunc)

```python
scaled_df = df_numeric * 10

print("\nScaled DataFrame (x10):\n", scaled_df.head())
```

 STRUCTURAL PRESERVATION CHECK

```python
print("\nIndex preserved in Series:", log_series.index.equals(sepal_length.index))
print("Columns preserved in DataFrame:", log_df.columns.equals(df_numeric.columns))
```

## Key Insight:

- NumPy ufuncs operate element-wise on Pandas objects  
- Index and column labels are preserved automatically  
- Output remains structurally aligned with input  
- Enables vectorized computation without losing data context  

##1.4.2 UFuncs: Index Alignment

When performing binary operations between two Series or DataFrames, Pandas automatically aligns data using index labels before computing results. This makes operations robust when datasets are incomplete or partially overlapping.

---

### Index alignment in Series

When two Series share only some common labels, Pandas matches values by index. Entries without a match produce missing values (NaN). The result contains the union of both index sets.

If needed, missing values can be handled explicitly using methods that allow a fill value instead of default NaN behavior.

---

### Index alignment in DataFrame

The same alignment logic extends to two-dimensional DataFrames. Both row and column labels are matched before performing operations.

If column or row labels differ between DataFrames, Pandas aligns them automatically and fills missing positions with NaN.

To control missing data behavior, operations can include a fill value (such as a scalar or computed statistic), which is used wherever alignment fails.

---

### Summary of behavior

- Alignment is label-based, not position-based  
- Missing matches result in NaN by default  
- `fill_value` can be used to replace missing entries  
- Both Series and DataFrame follow the same alignment rules  

This design ensures safe and consistent arithmetic across heterogeneous datasets.

 SERIES INDEX ALIGNMENT (PARTIAL OVERLAP)

```python
s1 = pd.Series(df["sepal length (cm)"].values, index=df.index)
s2 = pd.Series(df["sepal width (cm)"].values, index=df.index)
```

```python
# Create partial misalignment by subsetting indices
s1_partial = s1.iloc[:100]
s2_partial = s2.iloc[50:150]

print("Series 1 (partial):\n", s1_partial.head())
print("\nSeries 2 (partial):\n", s2_partial.head())
```

```python
# Alignment happens by index labels
aligned_sum = s1_partial + s2_partial

print("\nAligned Series Sum (NaN appears where no match exists):\n", aligned_sum.head(10))
```

HANDLING MISSING VALUES WITH fill_value

```python
aligned_sum_fill = s1_partial.add(s2_partial, fill_value=0)

print("\nAligned Series with fill_value=0:\n", aligned_sum_fill.head(10))
```

DATAFRAME INDEX ALIGNMENT (ROWS + COLUMNS)

```python
df1 = df.iloc[:100, :3]  # subset columns
df2 = df.iloc[50:150, 1:4]  # shifted rows + different columns

print("\nDataFrame 1 shape:", df1.shape)
print("DataFrame 2 shape:", df2.shape)

df_sum = df1 + df2

print("\nDataFrame aligned addition (NaN where mismatch occurs):\n", df_sum.head())
```

DATAFRAME ALIGNMENT WITH fill_value (using add)

```python
df_sum_fill = df1.add(df2, fill_value=0)

print("\nDataFrame with fill_value=0:\n", df_sum_fill.head())
```

- Pandas aligns data by LABELS, not positions  
- Series: union of indices with NaN for mismatches  
- DataFrame: alignment across both rows and columns  
- `fill_value` allows controlled replacement of missing matches  
- Ensures safe arithmetic on heterogeneous or incomplete datasets  

##1.4.3 Ufuncs: Operations Between DataFrame and Series

When performing operations between a DataFrame and a Series, Pandas applies broadcasting rules similar to NumPy, while also preserving label alignment.

By default, operations are applied row-wise, meaning each row of the DataFrame is aligned with the Series and operated on element by element.

If column-wise behavior is desired, Pandas allows explicit control using axis parameters, enabling operations along columns instead of rows.

Crucially, Pandas aligns indices during these operations. This means that if a Series has only a subset of columns or rows from a DataFrame, only matching labels are operated on, while non-matching entries produce missing values (NaN).

This combination of broadcasting and label alignment ensures that computations remain both flexible and context-aware, reducing errors when working with partially overlapping datasets.

CREATE A SERIES (used for broadcasting)

```python
feature_means = df.iloc[:, :4].mean()

print("\nFeature means (Series):\n", feature_means)
```

ROW-WISE BROADCASTING (DEFAULT BEHAVIOR)

```python
row_centered = df.iloc[:, :4] - feature_means

print("\nRow-wise broadcasting (first 5 rows):\n", row_centered.head())
```

 COLUMN-WISE BROADCASTING (EXPLICIT AXIS CONTROL)

```python
# subtract row-wise means (axis=1)
row_means = df.iloc[:, :4].mean(axis=1)

column_centered = df.iloc[:, :4].sub(row_means, axis=0)

print("\nColumn-wise broadcasting (first 5 rows):\n", column_centered.head())
```

PARTIAL INDEX ALIGNMENT BETWEEN SERIES AND DATAFRAME

```python
partial_series = pd.Series(
    feature_means.iloc[:2],  # only 2 matching columns
    index=feature_means.index[:2]
)

aligned_result = df.iloc[:, :4] + partial_series

print("\nPartial Series alignment (NaN expected):\n", aligned_result.head())
```

EXPLICIT AXIS CONTROL WITH SERIES

```python
axis_result = df.iloc[:, :4].add(feature_means, axis=1)

print("\nAxis=1 controlled broadcasting (first 5 rows):\n", axis_result.head())
```

- DataFrame-Series operations combine broadcasting + label alignment  
- Default behavior is row-wise alignment  
- `axis` parameter controls direction of operation  
- Missing labels result in NaN due to alignment rules  
- Enables safe vectorized computation on structured datasets  

##1.5 Handling Missing Data

Real-world datasets are often incomplete, inconsistent, or heterogeneous, unlike clean examples typically used in tutorials. A common challenge is the presence of missing values, which may be represented in different ways depending on the data source.

In Pandas, missing data is generally represented using standardized markers such as NaN (Not a Number) or NA. This allows consistent handling across numerical and non-numerical datasets.

The section introduces how Pandas represents missing values internally and how it provides built-in tools to detect, manage, and operate on such values in a structured and controlled way.

##1.5.1 Trade-Offs in Missing Data Conventions

Missing data is handled mainly in two ways:

Masking approach: uses a separate Boolean mask to mark missing values. This preserves data but increases memory and computation cost.

Sentinel approach: uses special values like NaN or -9999 to represent missing data. This is simpler but can limit valid value ranges and requires special handling in computations.

Different systems choose different strategies depending on efficiency and data type support.

MASKING APPROACH (FIXED)

```python
data = df.iloc[:, :4]  # isolate numeric features

# IMPORTANT: mask must match data shape exactly
mask = np.zeros(data.shape, dtype=bool)

# artificially mark missing values
mask[0:10, 0] = True
mask[20:30, 1] = True

masked_values = data.mask(mask)
print(masked_values.head(15))
```

 SENTINEL APPROACH (NaN)

```python
sentinel_df = data.copy()
sentinel_df.iloc[0:10, 0] = np.nan
sentinel_df.iloc[20:30, 1] = np.nan

print("\n---- SENTINEL (NaN) APPROACH ----")
print(sentinel_df.head(15))
```

ALT SENTINEL APPROACH (-9999)

```python
sentinel_alt = data.copy()
sentinel_alt.iloc[0:10, 0] = -9999
sentinel_alt.iloc[20:30, 1] = -9999
print(sentinel_alt.head(15))
```

COMPUTATIONAL DIFFERENCE

```python
print("\n---- COMPUTATION IMPACT ----")

print("Masked mean (NaN-safe):")
print(masked_values.mean())

print("\nSentinel mean (after cleaning required):")
print(sentinel_alt.replace(-9999, np.nan).mean())
```

- Mask must match EXACT shape of target DataFrame  
- Masking = explicit missing structure (safe but memory heavy)  
- Sentinel = embedded missing values (fast but error-prone)  
- NaN is the standard scientific computing convention in Pandas/NumPy  

##1.5.2 Missing Data in Pandas

Pandas handles missing data through constraints imposed by NumPy, which lacks a universal NA system for non-floating types. Designing explicit NA representations for every NumPy dtype (as in R) would be computationally inefficient and overly complex due to NumPy’s wide range of integer and floating variants.

An alternative based on masked arrays was also rejected because it introduces extra memory overhead and computational complexity.

---

## Design Choice

Pandas instead uses sentinel values:

- `None` (Python object-based missing value)  
- `NaN` (IEEE floating-point missing value)  

---

## None: Pythonic missing data

- Works only in object arrays  
- Forces Python-level computation (slow)  
- Breaks numerical operations (e.g., aggregation errors)  

---

## NaN: Numerical missing data

- Stored in `float64` arrays  
- Fast and supported at NumPy level  
- Propagates through all arithmetic operations  
- Requires special functions (`np.nansum`, etc.) to ignore  

---

## Pandas behavior

- Treats `None` and `NaN` interchangeably  
- Automatically converts:
  - integers → floats when NA appears  
  - `None` → `NaN` in numeric contexts  
- Ensures consistency but may change data types during processing  

---

Pandas prioritizes performance and simplicity, using sentinel-based missing data handling with automatic type casting rather than complex per-type NA encodings.

```python
# Keep only numeric feature block for clarity
data = df.iloc[:, :4].copy()
```

None (Python object missing data)

```python
none_df = data.copy()

# inject Python None values (forces object dtype behavior)
none_df.iloc[0:5, 0] = None
none_df.iloc[10:15, 1] = None

print("\n---- NONE (object-based missing data) ----")
print(none_df.head(10))
print("\nDtype check (None causes object conversion):")
print(none_df.dtypes)
```

NaN (IEEE floating-point missing data)

```python
nan_df = data.copy()

nan_df.iloc[0:5, 0] = np.nan
nan_df.iloc[10:15, 1] = np.nan

print("\n---- NaN (floating-point missing data) ----")
print(nan_df.head(10))
print("\nDtype check (NaN preserves float dtype):")
print(nan_df.dtypes)
```

COMPUTATIONAL BEHAVIOR COMPARISON

```python
# None-based data → requires object handling, often slower or breaks numerics
print("None-based mean (may fail or degrade):")
try:
    print(none_df.mean(numeric_only=True))
except Exception as e:
    print("Error:", e)

# NaN-based data → NumPy-compatible missing handling
print("\nNaN-based mean (automatic propagation):")
print(nan_df.mean())

print("\nNaN-safe aggregation:")
print(np.nansum(nan_df.to_numpy()))
```

TYPE CONVERSION EFFECT (PANDAS BEHAVIOR)

```python
mixed = data.copy()
mixed.iloc[0, 0] = None  # triggers upcasting
print(mixed.head())
print("\nDtypes after inserting None:")
print(mixed.dtypes)
```

- Pandas uses sentinel-based missing data (None, NaN)  
- None forces object dtype → slower computation  
- NaN integrates with NumPy float64 → efficient processing  
- Missing values trigger automatic type promotion  
- Design prioritizes performance over full per-type NA system  

##1.5.3 Operating on Null Values

Pandas treats `None` and `NaN` similarly and provides built-in tools to manage missing data:

- `isnull()` → identifies missing values (Boolean mask)  
- `notnull()` → identifies non-missing values  
- `dropna()` → removes missing data  
- `fillna()` → replaces or imputes missing values  

These functions enable efficient detection, filtering, and handling of null data.

## Detecting Null Values

Pandas provides:

- `isnull()` → returns Boolean mask of missing values  
- `notnull()` → returns Boolean mask of non-missing values  

These masks can be used for filtering data, e.g., selecting only non-null entries.

## Dropping Null Values

Use `dropna()` to remove NA values and `fillna()` to replace them.

### For Series:
- Simply removes null entries  

### For DataFrame:
- Drops entire rows or columns (not individual cells)  
- Default: drops rows with any null value (`how='any'`)  
- `axis=1`: drops columns with nulls  

### More control:
- `how='all'` → drop only rows/columns where all values are null  
- `thresh` → keep rows/columns with a minimum number of non-null values  

Overall: `dropna()` provides flexible control over removing missing data.

## Filling Null Values

`fillna()` replaces missing values instead of dropping them.

- Can use a constant value (e.g., 0) to fill all nulls  
- Supports forward-fill (`ffill`): fills using previous value  
- Supports backward-fill (`bfill`): fills using next value  
- For DataFrames, filling can be done along rows or columns using `axis`  
- If no prior (or next) value exists during fill, the null remains

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df["target"] = iris.target
```

DROPPING NULL VALUES

```python
# --- Series example ---
series_example = data["sepal length (cm)"]

series_dropped = series_example.dropna()
series_dropped
```

```python
# --- DataFrame examples ---

# Drop rows with ANY nulls (default)
df_drop_any = data.dropna()

# Drop rows only if ALL values are null
df_drop_all = data.dropna(how='all')

# Drop columns containing any null values
df_drop_cols = data.dropna(axis=1)

# Keep rows with at least 4 non-null values
df_thresh = data.dropna(thresh=4)

print("\n---- DATAFRAME dropna() ----")
print("Drop rows (any nulls):", df_drop_any.shape)
print("Drop rows (all nulls):", df_drop_all.shape)
print("Drop columns with nulls:", df_drop_cols.columns.tolist())
print("Threshold (≥4 non-null values):", df_thresh.shape)
```
