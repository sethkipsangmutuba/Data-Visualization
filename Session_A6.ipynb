##6.2) Comparison Operators as ufuncs (Brief Notes)

NumPy implements comparison operators as element-wise ufuncs.

Operations like `<`, `>`, `<=`, `>=`, `==`, `!=` return Boolean arrays.

Each element is compared individually against a value or another array.

Results are arrays of `True` and `False`.

Comparisons can be combined into compound expressions.

Internally, operators map to functions (e.g., `<` → `np.less`).

Works on arrays of any size and dimension.

Key idea: comparisons produce Boolean arrays used for filtering and analysis.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()


# 1. Basic element-wise comparisons

sepal_length = X[:, 0]

lt = sepal_length < 5.0
gt = sepal_length > 6.0
eq = sepal_length == 5.1

print("Less than 5.0 (first 10):", lt[:10])
print("Greater than 6.0 (first 10):", gt[:10])
print("Equal to 5.1 (first 10):", eq[:10])
```

```python
# 2. Comparison between arrays

petal_length = X[:, 2]

array_compare = sepal_length > petal_length

print("\nSepal > Petal (first 10):", array_compare[:10])
```

```python
# 3. Compound logical structure (vectorized logic)

compound_mask = (sepal_length > 5.0) & (petal_length < 4.0)

print("\nCompound condition (first 10):", compound_mask[:10])
```

```python
# 4. Counting using comparison results

count_condition = np.sum(compound_mask)

print("\nCount satisfying condition:", count_condition)
```

##6.3) Working with Boolean Arrays

Boolean arrays enable efficient counting, checking, and filtering operations.

---

### Counting entries

- `count_nonzero` counts `True` values directly.  
- `sum` also works (`True = 1`, `False = 0`).  
- Aggregation can be applied across rows or columns using `axis`.

---

### Logical checks

- `any()` checks if at least one `True` exists.  
- `all()` checks if all values are `True`.  
- Can be applied globally or along specific axes.

---

### Boolean operators

Combine conditions using:

- `&` (AND)  
- `|` (OR)  
- `^` (XOR)  
- `~` (NOT)  

Must use parentheses due to operator precedence.

Enables complex condition-based filtering.

---

### Key idea

Boolean arrays + aggregation = fast logical analysis without loops.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()

sepal_length = X[:, 0]
petal_length = X[:, 2]

# 1. Boolean conditions

mask1 = sepal_length > 5.0
mask2 = petal_length < 4.0
```

```python
# 2. Counting entries

count_true = np.count_nonzero(mask1)
count_sum = np.sum(mask1)

print("Count (nonzero):", count_true)
print("Count (sum):", count_sum)
```

```python
# 3. Axis-based aggregation

X_bool = X > 5.0

col_true = np.sum(X_bool, axis=0)
row_true = np.sum(X_bool, axis=1)

print("\nColumn-wise True counts:", col_true)
print("Row-wise True counts (first 5):", row_true[:5])
```

```python
# 4. Logical checks

any_true = np.any(mask1)
all_true = np.all(mask1)

print("\nAny True:", any_true)
print("All True:", all_true)
```

```python
# 5. Boolean operators (AND, OR, XOR, NOT)

and_mask = (sepal_length > 5.0) & (petal_length < 4.0)
or_mask = (sepal_length > 6.0) | (petal_length < 3.0)
xor_mask = (sepal_length > 5.5) ^ (petal_length < 4.0)
not_mask = ~(sepal_length > 5.0)

print("\nAND count:", np.sum(and_mask))
print("OR count:", np.sum(or_mask))
print("XOR count:", np.sum(xor_mask))
print("NOT count:", np.sum(not_mask))
```

##6.5) Boolean Arrays as Masks (Brief Notes)

Boolean masks are used to filter data based on conditions.

A condition like `x < 5` produces a Boolean array.

This Boolean array can be used to index the original array.

Only values where the mask is `True` are returned.

The result is a 1D array of selected elements.

---

## Applications

Masks allow extraction of subsets of data easily.

Can be combined with logical operators for complex filtering.

Useful for statistical analysis on specific conditions (e.g., rainy days).

Enables comparison between different groups within a dataset.

---

## Key idea

Boolean masking is a powerful way to select and analyze data efficiently without loops.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()

sepal_length = X[:, 0]
petal_length = X[:, 2]

# 1. Simple Boolean mask

mask = sepal_length < 5.5

filtered = sepal_length[mask]

print("Filtered values (first 10):", filtered[:10])
print("Filtered shape:", filtered.shape)
```

```python
# 2. Mask applied to full dataset (row selection)

filtered_rows = X[mask]

print("\nFiltered dataset shape:", filtered_rows.shape)
```

```python
# 3. Complex masking with logical operators

complex_mask = (sepal_length > 5.0) & (petal_length < 4.0)

subset = X[complex_mask]

print("\nComplex subset shape:", subset.shape)
```

```python
# 4. Statistical analysis on masked data

mean_subset = np.mean(subset, axis=0)

print("\nMean of filtered subset:", mean_subset)
```

```python
# 5. Comparison between groups

group1 = X[sepal_length < 5.5]
group2 = X[sepal_length >= 5.5]

print("\nGroup 1 mean (first feature):", np.mean(group1[:, 0]))
print("Group 2 mean (first feature):", np.mean(group2[:, 0]))
```
