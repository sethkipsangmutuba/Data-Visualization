##1.3 GroupBy: Split, Apply, Combine

Purpose: Conditional aggregation based on labels or index.  
Method: `groupby` operation in Pandas.  
Origin: Similar to SQL “GROUP BY”; concept popularized as “split-apply-combine” by Hadley Wickham.  

Idea:
- Split data into groups by key  
- Apply operations within each group  
- Combine results into final output  

##1.3.1 Split, Apply, Combine

Example idea: Summation is a typical “apply” operation in split-apply-combine.

---

### Split step:
Data is divided into groups based on a key column or label.

---

### Apply step:
A function is applied within each group (commonly aggregation, transformation, or filtering).

---

### Combine step:
Results from each group are merged into a single output structure.

---

## Key insight:
This process can be done efficiently without explicitly creating separate subsets of data.

---

## GroupBy mechanism:
Pandas internally handles grouping in a single pass, updating aggregates like sum, mean, count, or min per group.

---

## Important concept:
`groupby` does not immediately compute results; it creates a lazy GroupBy object that waits for an operation.

---

## Outcome:
Once an aggregation (e.g., sum) is applied, grouped results are computed and returned as a summarized dataset.

LOAD DATA

```python
import numpy as np
import pandas as pd
import seaborn as sns

planets = sns.load_dataset("planets")

# clean minimal subset for stable grouping
planets = planets.dropna(subset=["method", "year", "orbital_period"])
planets
```

SPLIT → APPLY → COMBINE (EXPLICIT DEMONSTRATION)

```python
# SPLIT (group formation)
groups = planets.groupby("method")
groups
```

```python
# APPLY (aggregation step)
sum_per_group = groups["orbital_period"].sum()
sum_per_group
```

```python
# COMBINE (final structured output)
result = sum_per_group.reset_index(name="total_orbital_period")
result
```

INTERNAL MECHANISM INTUITION (NO EXPLICIT SPLITTING)

```python
manual_check = (
    planets
    .groupby("method")["orbital_period"]
    .mean()
)
manual_check
```

LAZY EVALUATION DEMONSTRATION

```python
lazy_object = planets.groupby("method")
lazy_object
```

```python
# nothing computed until an operation is invoked
computed_count = lazy_object.size()
computed_count
```

- Split: data is logically partitioned by key (not physically separated)  
- Apply: computation happens within each group  
- Combine: results are merged into a structured output  
- GroupBy is lazy: no computation until aggregation is triggered  
- Pandas performs grouping in optimized internal passes rather than manual loops  

##1.3.2 The GroupBy Object

The GroupBy object is a flexible abstraction that behaves like a collection of grouped DataFrames, while deferring computation until an operation is applied.

Key idea: it enables efficient grouped analysis without manually splitting data.

---

## Core functionality

The most important GroupBy operations are:

- aggregation  
- filtering  
- transformation  
- apply  

These allow computation on subsets of data before recombining results.

---

## Column indexing

You can select specific columns within a GroupBy object. This narrows the focus to a Series-level grouping, but still delays computation until an operation (like median) is applied.

---

## Iteration over groups

GroupBy supports looping over groups, where each group is returned as a separate DataFrame or Series. This is useful for inspection or custom processing, though less efficient than built-in methods.

---

## Dispatch methods

Many standard DataFrame/Series methods automatically work on GroupBy objects. They are applied independently to each group, and results are recombined.

A common example is descriptive statistics per group, which helps reveal distributional differences across categories (e.g., discovery methods in the Planets dataset).

---

## Key insight

GroupBy applies operations per group automatically and efficiently, without requiring manual splitting. Any valid DataFrame or Series method can often be used, making it a powerful tool for structured data analysis.

##1.3.3. Aggregate, Filter, Transform, Apply (Reduced Notes)

GroupBy objects support four key operations before combining results: `aggregate()`, `filter()`, `transform()`, and `apply()`. These extend basic aggregation into more flexible group-level computations.

---

## Aggregation

Aggregation summarizes each group into reduced statistics (e.g., min, median, max). It can:

- apply multiple functions at once  
- target specific columns using mappings  

This produces compact summaries of each group.

---

## Filtering

Filtering removes entire groups based on a condition.

- A function is applied to each group  
- It returns True/False  
- Only groups passing the condition are kept  

Example idea: keep groups whose variability exceeds a threshold.

---

## Transformation

Transformation keeps the same shape as the original data.

- Operates group-wise  
- Returns transformed values aligned with original structure  
- Common use: centering or standardizing data within groups (e.g., subtracting group mean)  

Unlike aggregation, it does not reduce data size.

---

## Apply

Apply is the most flexible operation.

- Takes each group as a DataFrame  
- Returns either:
  - a DataFrame/Series, or  
  - a scalar value  
- Allows fully custom logic per group  

Example use: normalize one column using another group-wise statistic.

---

## Key takeaway

- Aggregate → reduces data  
- Filter → removes groups  
- Transform → keeps shape, modifies values  
- Apply → fully flexible custom computation  

Together, these make GroupBy a powerful framework for structured, group-level data analysis.

 LOAD DATASET

```python
import numpy as np
import pandas as pd
import seaborn as sns

planets = sns.load_dataset("planets")
planets = planets.dropna(subset=["method", "orbital_period", "year"]).copy()

grouped = planets.groupby("method")
grouped
```

AGGREGATION (REDUCTION PER GROUP)

AGGREGATION (SINGLE COLUMN)

```python
agg_result = grouped["orbital_period"].agg(["min", "median", "max"])
agg_result
```

AGGREGATION (MULTI COLUMN)

```python
agg_multi = grouped.agg({
    "orbital_period": ["mean", "std"],
    "year": "count"
})
agg_multi
```

FILTERING (GROUP SELECTION LOGIC)

```python
filtered = grouped.filter(lambda g: g["orbital_period"].std() > 1000)

print("\n--- FILTERED GROUPS (STD > 1000)---")
filtered["method"].value_counts()
```

TRANSFORMATION (SHAPE PRESERVING)

```python
planets["centered_period"] = grouped["orbital_period"].transform(
    lambda x: x - x.mean()
)

planets["standardized_period"] = grouped["orbital_period"].transform(
    lambda x: (x - x.mean()) / x.std()
)

print("\n--- TRANSFORMATION SAMPLE ---")
planets[["method", "orbital_period", "centered_period", "standardized_period"]]
```

 APPLY (FULL FLEXIBILITY PER GROUP)

```python
def custom_scaling(group):
    x = group["orbital_period"]
    return (x - x.min()) / (x.max() - x.min())

applied = grouped.apply(custom_scaling)
applied
```

- Aggregate → compresses each group into summary statistics
- Filter → removes entire groups based on logical condition
- Transform → preserves original shape while modifying values
- Apply → fully flexible, allows arbitrary group-level computation
- Together → constitute the full GroupBy analytical framework

##1.3.4 Specifying the Split Key (Reduced Notes)

GroupBy does not have to rely on a single column. The grouping key can be defined in multiple flexible ways.

---

## List / array / Series as keys

You can pass an external sequence of labels (same length as the DataFrame) to define groups.  
This allows grouping independent of existing columns.

---

## Column-based grouping (standard approach)

Grouping by a column is essentially equivalent to supplying that column as a key array.  
This is the most common and direct method.

---

## Dictionary or Series mapping

A mapping can assign each index value to a group label.

- Useful when categories are not directly stored in the data  
- Example idea: mapping categories like “A → vowel group”, “B/C → consonant group”  

This enables semantic grouping beyond raw data structure.

---

## Python function as key

A function can be applied to index values to determine grouping.

- Flexible and dynamic  
- Example: converting labels to lowercase or applying custom logic  

---

## Multiple keys (multi-level grouping)

You can combine multiple grouping strategies at once.

- Allows hierarchical grouping  
- Produces multi-index style grouped results  
- Useful for more granular aggregation patterns  

---

## Key takeaway

GroupBy supports data-driven, mapping-driven, function-driven, and multi-key grouping, making it highly adaptable for complex analytical structures beyond simple column-based splits.

LOAD DATASET

```python
import numpy as np
import pandas as pd
import seaborn as sns

planets = sns.load_dataset("planets").dropna(subset=["method", "orbital_period", "year"]).copy()
```

 LIST / ARRAY / SERIES AS KEYS (EXTERNAL GROUPING)

```python
external_keys = np.where(planets["orbital_period"] > 1000, "long_period", "short_period")

group_external = planets.groupby(external_keys)["orbital_period"].mean()
group_external
```

COLUMN-BASED GROUPING (STANDARD)

```python
group_method = planets.groupby("method")["orbital_period"].mean()
group_method
```

DICTIONARY / SERIES MAPPING

```python
method_map = {
    "Radial Velocity": "RV",
    "Transit": "Transit-Based",
    "Imaging": "Direct Imaging",
    "Microlensing": "Gravitational",
    "Eclipse Timing Variations": "Timing"
}

mapped_groups = planets["method"].map(method_map)
mapped_groups
```

```python
group_mapped = planets.groupby(mapped_groups)["orbital_period"].median()
group_mapped
```

FUNCTION-BASED GROUPING

```python
group_func = planets.groupby(lambda idx: len(str(idx)))["orbital_period"].mean()
group_func
```

MULTIPLE KEYS (HIERARCHICAL GROUPING)

```python
planets["decade"] = (planets["year"] // 10) * 10

multi_group = planets.groupby(["method", "decade"])["orbital_period"].mean()
multi_group
```

- Grouping is not restricted to columns

- This enables flexible and semantically rich data partitioning

##2. Pivot Tables

Pivot tables extend the idea of GroupBy to two-dimensional summarization.

- They take column-based data and reorganize it into a matrix (rows × columns)  
- Provide multidimensional aggregation in a single view  

---

## Relationship to GroupBy

- Both follow split–apply–combine logic  
- GroupBy → operates along a single dimension  
- Pivot tables → operate across a two-dimensional grid  

---

## Key idea

Pivot tables are essentially a multidimensional version of GroupBy, enabling more compact and intuitive summaries of relationships within data.

##2.1 Motivating Pivot Tables

Pivot tables provide a clearer way to explore multidimensional relationships in data, using structured summaries instead of long GroupBy pipelines.

---

## Dataset context

The Titanic dataset contains rich passenger-level data (e.g., sex, age, class, fare, survival), making it ideal for demonstrating grouped analysis.

---

##2.2 Pivot Tables by Hand (via GroupBy)

Basic insight: survival differs strongly by gender  
→ females had much higher survival rates than males  

Extending analysis:
- Grouping by sex + class reveals combined effects  
- Requires chaining operations (groupby → aggregate → unstack)  
- Works, but becomes harder to read  

---

##2.3 Pivot Table Syntax

Pivot tables simplify this process:

- Replace complex GroupBy pipelines with a single, readable operation  
- Automatically produce 2D summaries (rows × columns)  

Example structure:
- rows → one variable (e.g., sex)  
- columns → another variable (e.g., class)  
- values → aggregated metric (e.g., survival rate)  

Key insight: survival improves with both being female and higher class.

---

##2.4 Multilevel Pivot Tables

Pivot tables can include more dimensions:

- Add variables (e.g., age groups using binning)  
- Include additional column groupings (e.g., fare categories)  
- Results form hierarchical (multi-index) tables  

This enables 3D or 4D summaries within a structured grid.

---

##2.5 Additional Options

Pivot tables are flexible via parameters:

- `aggfunc` → controls aggregation type (mean, sum, etc.)  
- dictionary mapping → apply different functions to different columns  
- `margins=True` → adds totals (overall summaries across rows/columns)  
- `fill_value` / `dropna` → handle missing data  

---

## Key takeaway

Pivot tables provide a clean, readable, and scalable alternative to GroupBy for multidimensional analysis, especially when exploring interactions between multiple variables.

LOAD TITANIC DATASET

```python
import numpy as np
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
```

 DROP MISSING VALUES (CRITICAL PREPROCESSING)

```python
titanic_clean = titanic.dropna(subset=["sex", "class", "age", "fare", "survived"]).copy()
titanic_clean
```

GROUPBY (PIVOT BY HAND)

```python
# survival by sex
survival_sex = titanic_clean.groupby("sex")["survived"].mean()
survival_sex
```

```python
# survival by sex + class
survival_sex_class = (
    titanic_clean
    .groupby(["sex", "class"])["survived"]
    .mean()
    .unstack()
)
survival_sex_class
```

PIVOT TABLE (SIMPLIFIED APPROACH)

```python
pivot_basic = titanic_clean.pivot_table(
    values="survived",
    index="sex",
    columns="class",
    aggfunc="mean"
)

print("\n----- PIVOT TABLE (SEX × CLASS) ---")
pivot_basic
```

MULTILEVEL PIVOT TABLE (ADDING AGE GROUPS)

```python
titanic_clean["age_group"] = pd.cut(
    titanic_clean["age"],
    bins=[0, 18, 40, 80],
    labels=["child", "adult", "senior"]
)
titanic_clean["age_group"]
```

```python
pivot_multi = titanic_clean.pivot_table(
    values="survived",
    index=["sex", "age_group"],
    columns="class",
    aggfunc="mean"
)
pivot_multi
```

ADDITIONAL DIMENSION (FARE GROUPS)

```python
titanic_clean["fare_group"] = pd.qcut(
    titanic_clean["fare"],
    3,
    labels=["low", "mid", "high"]
)
titanic_clean["fare_group"]
```

```python
pivot_advanced = titanic_clean.pivot_table(
    values="survived",
    index=["sex", "age_group"],
    columns=["class", "fare_group"],
    aggfunc="mean"
)
pivot_advanced
```

MULTIPLE AGGREGATIONS

```python
pivot_multi_agg = titanic_clean.pivot_table(
    values=["survived", "fare"],
    index="sex",
    columns="class",
    aggfunc={"survived": "mean", "fare": "median"}
)
pivot_multi_agg
```

 MARGINS (TOTALS)

```python
pivot_margins = titanic_clean.pivot_table(
    values="survived",
    index="sex",
    columns="class",
    aggfunc="mean",
    margins=True
)
pivot_margins
```

- Missing values removed before analysis ensures valid aggregation
- Pivot tables replace complex groupby pipelines with clear structure
- MultiIndex pivots enable higher-dimensional analysis
- Aggregation, margins, and binning enhance interpretability
