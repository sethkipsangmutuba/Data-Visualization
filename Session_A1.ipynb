##1)  Understanding Data Types in Python — Lecture Notes (No Code)

This section establishes a foundational distinction in computational thinking: how Python manages data internally compared to statically typed languages, and why this distinction is crucial for numerical computing and data science.

---

### a. Central Idea of the Section

The core purpose is to understand:

- How Python represents data in memory  
- Why its design favors flexibility over computational efficiency  

This becomes essential when later transitioning to high-performance numerical tools like NumPy.

---

### b. Static Typing vs Dynamic Typing

#### Static Typing (e.g., C, Java)

In statically typed systems:

- Every variable has an explicitly declared type  
- The type is fixed before execution  
- Memory layout is predictable and optimized  

This leads to:

- High computational efficiency  
- Low runtime overhead  
- Less flexibility  

#### Dynamic Typing (Python)

Python uses dynamic typing:

- Variables do not require explicit type declarations  
- Types are inferred during execution  
- A variable can change type over time  

This enables:

- High flexibility  
- Ease of use  
- Rapid prototyping  

But it introduces:

- Runtime type checking overhead  
- Additional memory cost per object  
- Reduced efficiency for large-scale numerical operations  

---

### c. Key Concept: Variables as Objects

A critical conceptual shift is required:

- In Python, variables are not raw storage locations—they reference full objects  

Each object includes:

- The actual value  
- Type information  
- Internal metadata used by the interpreter  

Thus, even a simple number is stored as a structured object, not a raw primitive value.

---

### d. Flexibility vs Performance Trade-off

Python’s design prioritizes usability:

- Variables can change type dynamically  
- Data structures can hold mixed types  

However, this flexibility comes at a cost:

- Every operation requires type resolution at runtime  
- Memory usage increases due to object overhead  
- Large-scale numerical computation becomes inefficient  

---

### e. Core Limitation Highlighted

The key limitation exposed in this section is:

- Python’s object-based design is not optimized for large-scale numerical computation  

This becomes especially visible when handling:

- Large datasets  
- Repeated numerical operations  
- High-dimensional arrays  

---

### f. Why This Matters for Data Science

In data science workflows, efficiency is critical. The overhead introduced by Python’s dynamic object system creates bottlenecks in:

- Statistical computation  
- Machine learning pipelines  
- Large-scale data processing  

---

### 7. Conceptual Transition to NumPy

This limitation motivates the need for NumPy, which introduces:

- Fixed-type numerical arrays  
- Contiguous memory storage  
- Reduced object overhead

##1.1)  A Python Integer Is More Than Just an Integer

In Python, an integer is not a simple raw value but a high-level object implemented through an underlying C structure.

It stores not only the numeric value but also metadata such as:
- reference counting (for memory management)
- type information
- object size
- internal digit representation of the integer value

This object-based design enables Python’s flexibility and dynamic behavior, allowing integers to seamlessly integrate into a unified object system.

However, this abstraction introduces both computational and memory overhead compared to low-level languages like C, where integers are stored as fixed-size primitive values with minimal metadata.

As a result, large-scale numerical computations involving many Python integers are less memory-efficient and slower in aggregate operations.

This limitation motivates the use of optimized array-based structures such as NumPy, where numerical values are stored in contiguous, fixed-type memory blocks with significantly reduced overhead.

```python
# Convert sklearn Bunch → DataFrame

iris_df = pd.DataFrame(
    data=iris.data,
    columns=iris.feature_names
)

# 1. Integer Representation

int_series = (iris_df.iloc[:, 0] * 100).astype(int)

int_list = int_series.tolist()
int_numpy = int_series.to_numpy()


# 2. Python Integer Object Inspection

x = int_list[0]

print("Value:", x)
print("Type:", type(x))
print("Memory size of Python int:", sys.getsizeof(x), "bytes")
```

```python
# 3. NumPy Integer Inspection

np_val = int_numpy[0]

print("\nNumPy value:", np_val)
print("NumPy dtype:", int_numpy.dtype)
print("Memory per element:", int_numpy.itemsize, "bytes")
```

```python
# 4. Memory Comparison

list_memory = sys.getsizeof(int_list) + sum(sys.getsizeof(i) for i in int_list)
numpy_memory = int_numpy.nbytes

print("\nMemory Usage:")
print("Python list:", list_memory, "bytes")
print("NumPy array:", numpy_memory, "bytes")
```

```python
# 5. Computation Timing

def sum_list(data):
    total = 0
    for v in data:
        total += v
    return total

start = time.time()
list_sum = sum_list(int_list)
list_time = time.time() - start

start = time.time()
numpy_sum = np.sum(int_numpy)
numpy_time = time.time() - start

print("\nSum (List):", list_sum)
print("Sum (NumPy):", numpy_sum)

print("\nExecution Time:")
print("List:", list_time)
print("NumPy:", numpy_time)
```

##1.2 A Python List Is More Than Just a List

A Python list is a flexible, mutable container that can store multiple Python objects of potentially different types. This flexibility arises from Python’s dynamic typing, where each element in a list is a full Python object carrying its own metadata such as type information and reference count.

While this design allows heterogeneous collections (for example integers, strings, and floats in the same list), it introduces overhead because each element is independently managed as a complete object.

Internally, a Python list does not store raw values directly; instead, it stores references (pointers) to separate objects distributed in memory.

This structure contrasts with fixed-type arrays, such as those in NumPy, which store data in a single contiguous memory block with uniform type information. As a result, lists prioritize flexibility, while fixed-type arrays prioritize efficiency in memory usage and computational speed especially important in large scale data science and numerical computation contexts.

```python
import numpy as np
import pandas as pd

# Create a heterogeneous Python list from Iris data
# (int, float, string mixed structure to show flexibility)

python_list = [
    iris_df.iloc[0, 0],              # float
    int(iris_df.iloc[0, 1] * 100),   # int
    "iris_sample",                   # string
    iris_df.iloc[0, 2]               # float
]


# 1. Inspect List Structure

print("Python List:", python_list)

for i, item in enumerate(python_list):
    print(f"Element {i}: value={item}, type={type(item)}, size={sys.getsizeof(item)} bytes")
```

```python
# 2. Memory of List Container vs Elements

list_container_size = sys.getsizeof(python_list)
elements_size = sum(sys.getsizeof(item) for item in python_list)

print("\nList container size:", list_container_size, "bytes")
print("Sum of element sizes:", elements_size, "bytes")
```

```python
# 3. Contrast with NumPy Array (fixed-type)

numeric_array = iris_df.iloc[:, 0].to_numpy()

print("\nNumPy dtype:", numeric_array.dtype)
print("NumPy itemsize:", numeric_array.itemsize, "bytes per element")
print("Total NumPy memory:", numeric_array.nbytes, "bytes")
```

```python
# 4. Demonstrate Heterogeneity vs Homogeneity

try:
    # This works for list (heterogeneous allowed)
    python_list.append({"new_type": "dictionary"})
    print("\nList after adding dict:", python_list[-1])
except Exception as e:
    print(e)

# NumPy strict typing behavior
try:
    numeric_array = np.append(numeric_array, "text")  # forces type upcast
    print("\nNumPy array after appending string:", numeric_array[-1])
    print("New dtype:", numeric_array.dtype)
except Exception as e:
    print(e)
```

##1.3) Fixed-Type Arrays in Python

Python provides multiple mechanisms for storing data in fixed-type, memory-efficient structures where all elements share the same data type. This design reduces overhead compared to standard Python lists, which store references to full Python objects.

One built-in option is the `array` module, which allows creation of dense arrays with a specified uniform type using type codes (for example, integers). These arrays are more memory-efficient than lists because they store raw values rather than object references.

However, their functionality is limited compared to NumPy arrays. The **NumPy ndarray** extends this idea significantly by combining fixed-type storage with highly optimized numerical operations, enabling fast computation over large datasets.

NumPy is typically imported under a standard alias for convenience, forming the foundation for most scientific computing workflows in Python.

```python
import array

numeric_col = iris_df.iloc[:, 0]

# 1. Python List (object-based)

py_list = numeric_col.tolist()
```

```python
# 2. Built-in array module (fixed-type)

# 'd' = double precision float
py_array = array.array('d', numeric_col)
```

```python
# 3. NumPy array (optimized fixed-type)

np_array = numeric_col.to_numpy()
```

```python
# 4. Memory Comparison

print("\n--- Memory Usage ---")

list_memory = sys.getsizeof(py_list) + sum(sys.getsizeof(x) for x in py_list)
array_memory = sys.getsizeof(py_array)
numpy_memory = np_array.nbytes

print("Python list:", list_memory, "bytes")
print("array module:", array_memory, "bytes")
print("NumPy array:", numpy_memory, "bytes")
```

```python
# 5. Type Uniformity Check

print("\n--- Type Inspection ---")

print("array module type:", py_array.typecode)
print("NumPy dtype:", np_array.dtype)

print("First element types:")
print("List:", type(py_list[0]))
print("array:", type(py_array[0]))
print("NumPy:", type(np_array[0]))
```

```python
# 6. Computation Performance

import time

def sum_list(data):
    total = 0
    for v in data:
        total += v
    return total

# List timing
start = time.time()
sum_list(py_list)
list_time = time.time() - start

# array module timing
start = time.time()
sum(py_array)
array_time = time.time() - start

# NumPy timing
start = time.time()
np.sum(np_array)
numpy_time = time.time() - start

print("\n--- Computation Time ---")
print("List:", list_time)
print("array module:", array_time)
print("NumPy:", numpy_time)
```

##1.4) Creating Arrays from Python Lists

NumPy arrays can be directly constructed from Python lists using a unified array constructor. This provides a bridge between Python’s flexible list structures and NumPy’s efficient fixed-type numerical arrays.

A key distinction is that NumPy arrays enforce homogeneity: all elements must share a single data type. When mixed types are provided, NumPy applies type upcasting, converting values into a common compatible type (for example, integers may be promoted to floating-point numbers).

If required, the data type can be explicitly defined using the `dtype` parameter, allowing precise control over memory representation and numerical precision.

Beyond one-dimensional structures, NumPy naturally supports multidimensional arrays. Nested Python lists are interpreted as rows in a matrix-like structure, enabling direct creation of 2D arrays without additional transformation. This forms the foundation for higher-dimensional data modeling in scientific computing and data science workflows.

```python
# 1. Creating NumPy Array from Python List

python_list = iris_df.iloc[:, 0].tolist()

np_array = np.array(python_list)

print("Original Python list (first 5):", python_list[:5])
print("NumPy array (first 5):", np_array[:5])
```

```python
# 2. Type Homogeneity + Upcasting

mixed_list = [1, 2, 3.5, 4, 5]

mixed_array = np.array(mixed_list)

print("\nMixed Python list:", mixed_list)
print("NumPy array:", mixed_array)
print("NumPy dtype (after upcasting):", mixed_array.dtype)
```

```python
# 3. Explicit dtype Control

int_array = np.array(mixed_list, dtype=np.int32)

print("\nExplicit dtype (int32):", int_array)
print("dtype:", int_array.dtype)

float_array = np.array(mixed_list, dtype=np.float64)

print("Explicit dtype (float64):", float_array)
print("dtype:", float_array.dtype)
```

```python
# 4. Multidimensional Arrays from Nested Lists

nested_list = iris_df.iloc[:5, :3].values.tolist()

matrix = np.array(nested_list)

print("\nNested Python list (first 2 rows):", nested_list[:2])
print("\nNumPy 2D array:\n", matrix)

print("\nShape:", matrix.shape)
print("Dimensionality:", matrix.ndim)
```

## Creating Arrays from Scratch

NumPy provides a rich set of built-in functions for constructing arrays directly from defined patterns, which is especially efficient for large scale data initialization and simulation tasks.

These constructors allow users to generate arrays with specific structures without manually populating elements:

**Constant arrays:** Arrays filled entirely with zeros, ones, or any fixed value are commonly used for initialization, baseline modeling, and placeholders.

**Sequential arrays:** Evenly spaced numeric sequences can be generated over a defined interval, supporting controlled sampling and feature generation.

**Linearly spaced arrays:** Values can be distributed uniformly between endpoints, which is particularly useful in numerical analysis and plotting.

**Random arrays:** NumPy supports both uniform and normal distributions, enabling stochastic simulation, Monte Carlo methods, and statistical modeling.

**Integer random arrays:** Useful for discrete simulations, sampling, and algorithm testing.

**Identity matrices:** Fundamental in linear algebra, representing neutral transformation operators.

**Uninitialized arrays:** Allocate memory without setting values, offering performance benefits when values will be overwritten immediately.

These methods highlight NumPy’s focus on computational efficiency and flexibility, allowing direct construction of structured numerical data essential for scientific computing workflows.

```python
# 1. Constant Arrays

zeros_array = np.zeros((3, 4))
ones_array = np.ones((3, 4))
full_array = np.full((3, 4), 7)

print("Zeros:\n", zeros_array)
print("\nOnes:\n", ones_array)
print("\nFull (7):\n", full_array)
```

```python
# 2. Sequential Arrays

seq_array = np.arange(0, 10, 2)

print("\nSequential (arange):", seq_array)
```

```python
# 3. Linearly Spaced Arrays

lin_array = np.linspace(0, 1, 5)

print("\nLinearly spaced (linspace):", lin_array)
```

```python
# 4. Random Arrays (Uniform & Normal)

uniform_random = np.random.rand(3, 3)
normal_random = np.random.randn(3, 3)

print("\nUniform random:\n", uniform_random)
print("\nNormal random:\n", normal_random)
```

```python
# 5. Integer Random Arrays

int_random = np.random.randint(0, 10, size=(3, 3))

print("\nInteger random:\n", int_random)
```

```python
# 6. Identity Matrix

identity = np.eye(4)

print("\nIdentity matrix:\n", identity)
```

```python

# 7. Uninitialized Array

empty_array = np.empty((3, 3))

print("\nUninitialized array:\n", empty_array)
```

##1.5) NumPy Standard Data Types

NumPy arrays are fundamentally homogeneous, meaning every element must share the same data type. This design ensures computational efficiency and predictable memory usage, but it also requires careful attention to data type selection.

Because NumPy is implemented in C, its data types closely mirror low-level machine representations, including fixed-size integers, floating-point numbers, unsigned integers, and complex numbers. These types differ in memory footprint and numerical range, which directly affects precision, performance, and storage efficiency.

When creating arrays, data types can be explicitly defined either through string identifiers or NumPy’s built-in type objects. This allows precise control over how data is stored and processed, which is especially important in large-scale numerical computation, statistical modeling, and scientific analysis.

More advanced type systems, including structured and compound types, extend NumPy’s capability beyond simple numerical arrays, forming a bridge toward relational and heterogeneous data representations used later in scientific workflows.

```python
import numpy as np

col = iris_df.iloc[:, 0]


# 1. Default dtype inference

default_array = np.array(col)

print("Default dtype:", default_array.dtype)
```

```python
# 2. Explicit numeric types

float32_array = np.array(col, dtype=np.float32)
float64_array = np.array(col, dtype=np.float64)

int32_array = np.array(col, dtype=np.int32)
uint8_array = np.array(col, dtype=np.uint8)

print("\nFloat32 dtype:", float32_array.dtype)
print("Float64 dtype:", float64_array.dtype)
print("Int32 dtype:", int32_array.dtype)
print("UInt8 dtype:", uint8_array.dtype)
```

```python
# 3. Memory footprint comparison

print("\n--- Memory per array ---")

print("float32 total bytes:", float32_array.nbytes)
print("float64 total bytes:", float64_array.nbytes)
print("int32 total bytes:", int32_array.nbytes)
print("uint8 total bytes:", uint8_array.nbytes)
```

```python
# 4. Precision consequences

print("\n--- Precision example ---")

print("Original value:", col.iloc[0])
print("float32:", float32_array[0])
print("float64:", float64_array[0])
print("int32:", int32_array[0])
```

```python
# 5. Complex numbers (extended type system)

complex_array = np.array(col, dtype=np.complex128)

print("\nComplex dtype:", complex_array.dtype)
print("Example value:", complex_array[0])
```

```python
# 6. Structured dtype

structured = np.array(
    list(zip(
        iris_df.iloc[:, 0],
        iris_df.iloc[:, 1]
    )),
    dtype=[('feature1', 'f8'), ('feature2', 'f8')]
)

print("\nStructured dtype:", structured.dtype)
print("First element:", structured[0])
```
