## Computation on Arrays: Broadcasting

Broadcasting is a method for vectorizing operations on arrays of different shapes.

It allows NumPy to apply arithmetic operations without explicit loops.

Works by automatically expanding smaller arrays to match larger ones.

Commonly used with binary operations like addition, subtraction, and multiplication.

Improves computational efficiency and code simplicity.

##5.1) Introducing Broadcasting (Brief Notes)

Broadcasting enables operations between arrays of different shapes.

Scalar values can be treated as arrays and applied element-wise.

Smaller arrays are conceptually “stretched” to match larger array shapes.

This stretching does not create actual copies in memory.

Works across dimensions, allowing interaction between 1D and 2D arrays.

Both arrays can be broadcast simultaneously to a common shape.

Results often produce higher-dimensional arrays depending on inputs.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()

# 1. Scalar broadcasting (0D → ND)

scalar_add = X + 2
scalar_mul = X * 10

print("Scalar broadcasting (X + 2):", scalar_add[0])
```

```python
# 2. 1D array broadcasted to 2D

row_vector = np.array([1, 2, 3, 4])

broadcast_row = X + row_vector  # row-wise broadcast

print("\nRow vector broadcast (first row):", broadcast_row[0])
```

```python
# 3. Column-wise broadcasting (reshape needed)

col_vector = np.array([1, 2, 3, 4]).reshape(4, 1)

broadcast_col = X.T + col_vector  # transpose for alignment

print("\nColumn vector broadcast (first column):", broadcast_col[:, 0])
```

```python
# 4. 1D vs 2D interaction (implicit expansion)

y = iris_df.iloc[:, 1].to_numpy()

broadcast_mix = X[:, 0] + y  # element-wise alignment

print("\n1D–1D broadcast result:", broadcast_mix[:5])
```

```python
# 5. Shape inspection

print("\nOriginal X shape:", X.shape)
print("Row vector shape:", row_vector.shape)
print("Column vector shape:", col_vector.shape)
print("Broadcast result shape:", broadcast_row.shape)
```

##5.2) Rules of Broadcasting (Brief Notes)

Broadcasting follows three core rules to align array shapes.

Rule 1: Arrays with fewer dimensions are padded with ones on the left.

Rule 2: Dimensions with size 1 are stretched to match the other array.

Rule 3: If dimensions differ and neither is 1, broadcasting fails.

Compatible arrays can be expanded to a common shape for operations.

Incompatible shapes raise an error due to mismatch.

Padding always occurs on the left, not the right.

Reshaping (e.g., adding a new axis) can make arrays compatible.

Rules apply to all binary operations, not just addition.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()


# RULE 1: Left padding with ones

a = np.array([1, 2, 3, 4])        # shape (4,)
b = X                             # shape (150, 4)

result_rule1 = b + a              # (4,) → (1,4) → broadcast

print("Rule 1 result shape:", result_rule1.shape)
```

```python
# RULE 2: Dimension of size 1 is stretched

c = np.array([[10], [20], [30], [40]])  # shape (4,1)

# align for broadcasting with X.T (4,150)
result_rule2 = X.T + c

print("Rule 2 result shape:", result_rule2.shape)
```

```python
# RULE 3: Incompatible shapes (should fail)

try:
    d = np.array([1, 2, 3])       # shape (3,)
    bad_result = X + d            # (150,4) vs (3,) → mismatch
except ValueError as e:
    print("\nRule 3 error (expected):", e)
```

```python
# RULE 4: Explicit reshaping to fix incompatibility

d_fixed = d.reshape(1, 3)         # make compatible via new axis

try:
    fixed_result = X[:, :3] + d_fixed
    print("\nFixed broadcasting shape:", fixed_result.shape)
except Exception as e:
    print(e)
```

```python
# RULE 5: Left-padding demonstration explicitly

e = np.array(5)                   # scalar → shape ()
result_rule5 = X + e

print("\nRule 5 scalar broadcast shape:", result_rule5.shape)
```

##5.3) Broadcasting in Practice

Broadcasting is widely used for efficient data transformations.

---

### Centering an array

Used to normalize data by subtracting the mean.

Mean is computed per feature (column-wise).

Broadcasting allows subtraction without loops.

Resulting array has mean approximately zero (within numerical precision).

---

### Plotting a two-dimensional function

Broadcasting enables evaluation of functions over a grid.

1D arrays are expanded to form a 2D coordinate system.

Useful for visualization of mathematical functions.

Produces structured arrays suitable for plotting and analysis.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()

# Compute column-wise mean

mean_vec = np.mean(X, axis=0)

# Centering using broadcasting

X_centered = X - mean_vec

print("Original column means:", mean_vec)
print("Centered means (approx 0):", np.mean(X_centered, axis=0))
```

```python
import numpy as np

# Create 1D coordinate ranges
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

# Broadcasting to create 2D grid

X_grid = x[:, np.newaxis]   # shape (100,1)
Y_grid = y[np.newaxis, :]   # shape (1,100)

# Evaluate 2D function using broadcasting
# Example: f(x,y) = sin(x) * cos(y)

Z = np.sin(X_grid) * np.cos(Y_grid)

print("Grid shape X:", X_grid.shape)
print("Grid shape Y:", Y_grid.shape)
print("Output shape Z:", Z.shape)
```
