## Statistical Framing

We model:

$$
tip = \beta_0 + \beta_1 \cdot total\_bill + \epsilon
$$

###Define Predictor and Response

Define the predictor (independent variable) and response (dependent variable):

$$
X_i = total\_bill_i, \quad i = 1, 2, \dots, n
$$

$$
Y_i = tip_i, \quad i = 1, 2, \dots, n
$$

where $n = 244$ observations in the dataset.

---

### Vector Form

$$
X =
\begin{bmatrix}
X_1 \\
X_2 \\
\vdots \\
X_n
\end{bmatrix},
\quad
Y =
\begin{bmatrix}
Y_1 \\
Y_2 \\
\vdots \\
Y_n
\end{bmatrix}
$$

```python
X = tips[["total_bill"]]  # Predictor as DataFrame
y = tips["tip"]           # Response
```

###Add Intercept Term

We introduce a column of ones to account for the intercept $\beta_0$:

$$
X_{design} =
\begin{bmatrix}
1 & X_1 \\
1 & X_2 \\
\vdots & \vdots \\
1 & X_n
\end{bmatrix}
=
\begin{bmatrix}
1 & total\_bill_1 \\
1 & total\_bill_2 \\
\vdots & \vdots \\
1 & total\_bill_n
\end{bmatrix}
$$

- **First column**: intercept ($\beta_0$)  
- **Second column**: predictor $X_i$  

### Dimension

$$
X_{design} \in \mathbb{R}^{n \times 2}, \quad n = 244
$$

### Role of the Intercept ($\beta_0$)

Including $\beta_0$ allows the regression line to have a **non-zero baseline tip** when the total bill is zero.  
- Mathematically required for correct linear modeling  
- Provides flexibility to shift the line vertically to best fit the data  
- While not practically meaningful in the restaurant context (a bill of $0$ is impossible), it ensures unbiased estimation of the slope $\beta_1$

```python
X = sm.add_constant(X)  # Adds beta_0 (intercept)
X.head()
```

###Fit the OLS Regression Model

We model the response $Y_i$ as a linear function of the predictor $X_i$ and an intercept $\beta_0$:

$$
Y_i = \beta_0 + \beta_1 X_i + \epsilon_i, \quad i = 1, 2, \dots, n
$$

Where:

- $\beta_0$ = intercept  
- $\beta_1$ = slope (effect of total bill on tip)  
- $\epsilon_i$ = error term, assumed i.i.d., $E[\epsilon_i] = 0$, $\text{Var}[\epsilon_i] = \sigma^2$  

---

### Matrix Form

$$
Y = X_{design} \beta + \epsilon
$$

Where:

$$
Y =
\begin{bmatrix}
Y_1 \\
Y_2 \\
\vdots \\
Y_n
\end{bmatrix}, \quad
X_{design} =
\begin{bmatrix}
1 & X_1 \\
1 & X_2 \\
\vdots & \vdots \\
1 & X_n
\end{bmatrix}, \quad
\beta =
\begin{bmatrix}
\beta_0 \\
\beta_1
\end{bmatrix}, \quad
\epsilon =
\begin{bmatrix}
\epsilon_1 \\
\epsilon_2 \\
\vdots \\
\epsilon_n
\end{bmatrix}
$$

---

### OLS Estimator

The fitted coefficients $\hat{\beta}$ are obtained by minimizing the sum of squared errors:

$$
\hat{\beta} = \arg \min_\beta \sum_{i=1}^{n} (Y_i - \beta_0 - \beta_1 X_i)^2
$$

Closed-form solution:

$$
\hat{\beta} = (X_{design}^\top X_{design})^{-1} X_{design}^\top Y
$$

Residuals:

$$
\hat{\epsilon} = Y - X_{design} \hat{\beta}
$$

```python
ols_model = sm.OLS(y, X).fit()
ols_model
```

###View Regression Summary

```python
print(ols_model.summary())
```

### Estimation of Coefficients

We estimate the coefficients $\hat{\beta}_0$ (intercept) and $\hat{\beta}_1$ (slope) that **minimize the sum of squared residuals**, i.e., they provide the **best linear fit** to the data:

$$
\hat{\beta} = \arg \min_\beta \sum_{i=1}^{n} (Y_i - \beta_0 - \beta_1 X_i)^2
$$

- $\hat{\beta}_0$ shifts the regression line vertically  
- $\hat{\beta}_1$ determines the slope, representing the effect of $total\_bill$ on $tip$  
- This is the standard **Ordinary Least Squares (OLS)** estimation

```python
intercept = ols_model.params["const"]
intercept
```

- Represents the expected tip when the total bill is zero.
- While a bill of zero is not practical, it sets the baseline level of the regression line.

```python
slope = ols_model.params["total_bill"]
slope
```

- Each 1-unit increase in total bill increases the expected tip by approximately **0.105 units**  
- In other words, the tip grows at roughly **10.5% of the total bill on average**

##95% Confidence Interval for Slope ($\beta_1$)

The 95% confidence interval (CI) for the slope $\beta_1$ is:

$$
CI_{95\%}(\beta_1) = \hat{\beta}_1 \pm t_{0.975,\, n-2} \cdot SE(\hat{\beta}_1)
$$

Where:

- $\hat{\beta}_1 = 0.1050$ → estimated slope  
- $SE(\hat{\beta}_1) = 0.007$ → standard error of the slope  
- $t_{0.975,\, n-2} \approx 1.97$ → critical value from $t$-distribution with $n-2 = 242$ degrees of freedom  

### Calculation

$$
0.1050 \pm 1.97 \cdot 0.007 = 0.1050 \pm 0.0138
$$

$$
CI_{95\%}(\beta_1) = [0.0912, 0.1188] \approx [0.091, 0.120]
$$

---

We are **95% confident** that the true effect of total bill on tip lies between **0.091 and 0.120**.  

- Practically: for every \$1 increase in bill, the tip increases by **9.1–12.0 cents** on average.

```python
ci = ols_model.conf_int().loc["total_bill"]
ci
```

##Model Fit Metrics

##R-squared ($R^2$)

The coefficient of determination $R^2$ is defined as:

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2}{\sum_{i=1}^{n} (Y_i - \bar{Y})^2}
$$

Where:

- $Y_i$ = observed tip  
- $\hat{Y}_i = \hat{\beta}_0 + \hat{\beta}_1 X_i$ = predicted tip  
- $\bar{Y} = \frac{1}{n} \sum_{i=1}^{n} Y_i$ = mean tip  

### Interpretation

- $R^2 = 0.4566$ → approximately **45.7% of the variance in tip** is explained by total bill

```python
r2 = ols_model.rsquared
r2
```

## Adjusted R-squared ($R_{adj}^2$)

The adjusted $R^2$ accounts for model complexity:

$$
R_{adj}^2 = 1 - (1 - R^2) \frac{n-1}{n-p-1}
$$

Where:

- $n = 244$ (observations)  
- $p = 1$ (number of predictors)  

### Interpretation

- $R_{adj}^2 = 0.4544$ → almost the same as $R^2$, confirming the model is **not overfitting**  
- In short: **total bill explains nearly half of the variability in tip**, representing a moderate fit

```python
adj_r2 = ols_model.rsquared_adj
adj_r2
```

##F-Statistic

The F-statistic in simple linear regression tests whether the model explains a **significant portion of the variance**:

$$
F = \frac{MS_{reg}}{MS_{res}} = \frac{SS_{reg}/p}{SS_{res}/(n-p-1)}
$$

Where:

- $SS_{reg} = \sum_{i=1}^{n} (\hat{Y}_i - \bar{Y})^2$ → regression sum of squares  
- $SS_{res} = \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2$ → residual sum of squares  
- $p = 1$ → number of predictors  
- $n = 244$ → number of observations  

### Interpretation

- $F = 203.435$ → the regression model is **highly significant**  
- Corresponding $p \approx 6.69 \times 10^{-34}$ → reject $H_0: \beta_1 = 0$  

**In simple terms:** total bill is a **strong and significant predictor** of tip

```python
f_stat = ols_model.fvalue
f_stat
```

## P-value for the F-Test

The p-value for the F-test is defined as:

$$
p_F = P(F_{p,\, n-p-1} \ge F_{obs})
$$

Where:

- $F_{obs} = 203.435$ → observed F-statistic  
- $F_{p,\, n-p-1}$ → F-distribution with $p = 1$ predictor and $n-p-1 = 242$ degrees of freedom  
- Null hypothesis: $H_0: \beta_1 = 0$ (predictor has no effect)  

###Interpretation

- $p \approx 6.69 \times 10^{-34}$ → extremely small, essentially zero  
- Strongly reject $H_0$  

**Meaning:** total bill is a **statistically significant predictor** of tip, and the regression explains a **meaningful portion of variance**

```python
f_pval = ols_model.f_pvalue
f_pval
```

##Residual Analysis (Normality)

## Residuals & Normality Tests

### Residuals

The residuals are defined as:

$$
\hat{\epsilon}_i = Y_i - \hat{Y}_i = Y_i - (\hat{\beta}_0 + \hat{\beta}_1 X_i), \quad i = 1, \dots, n
$$

Vector form:

$$
\hat{\epsilon} =
\begin{bmatrix}
\hat{\epsilon}_1 \\
\hat{\epsilon}_2 \\
\vdots \\
\hat{\epsilon}_n
\end{bmatrix}
$$

---

### Jarque-Bera Test

Test statistic:

$$
JB = \frac{n}{6} \left( S^2 + \frac{(K-3)^2}{4} \right)
$$

Where:

- $n = 244$ → sample size  
- $S = \text{skewness}(\hat{\epsilon}) = \frac{\frac{1}{n} \sum (\hat{\epsilon}_i - \bar{\hat{\epsilon}})^3}{\left(\frac{1}{n} \sum (\hat{\epsilon}_i - \bar{\hat{\epsilon}})^2 \right)^{3/2}}$  
- $K = \text{kurtosis}(\hat{\epsilon}) = \frac{\frac{1}{n} \sum (\hat{\epsilon}_i - \bar{\hat{\epsilon}})^4}{\left(\frac{1}{n} \sum (\hat{\epsilon}_i - \bar{\hat{\epsilon}})^2 \right)^{2}}$  

P-value:

$$
p_{JB} = P(\chi^2_2 \ge JB)
$$

- Null hypothesis: residuals are normally distributed  
- Alternative hypothesis: residuals are not normally distributed  

---

### Skew and Kurtosis

- Skew:  
$$
\text{Skew} = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{\hat{\epsilon}_i - \bar{\hat{\epsilon}}}{s_\epsilon} \right)^3
$$

- Kurtosis:  
$$
\text{Kurtosis} = \frac{1}{n} \sum_{i=1}^{n} \left( \frac{\hat{\epsilon}_i - \bar{\hat{\epsilon}}}{s_\epsilon} \right)^4
$$

Where $s_\epsilon^2 = \frac{1}{n} \sum (\hat{\epsilon}_i - \bar{\hat{\epsilon}})^2$ is the residual variance.

---

###  Interpretation

- Residual histogram → visually inspect normality and symmetry  
- Jarque-Bera test: $JB \approx 37.75$, $p \approx 6.35 \times 10^{-9}$ → reject null → residuals **not normally distributed**  
- Skew = 0.443 → slight right-skew of residuals  
- Kurtosis = 4.711 → heavier tails than normal (leptokurtic)  

**Interpretation:** While total bill predicts tip, other unobserved factors influence tipping, resulting in **non-normal residuals**.

```python
# Extract residuals
residuals = ols_model.resid

# Histogram of residuals
plt.figure(figsize=(8, 5))
sns.histplot(residuals, kde=True)
plt.title("Residuals Distribution")
plt.xlabel("Residuals")
plt.show()

# Jarque-Bera test
from scipy.stats import jarque_bera
jb_stat, jb_p = jarque_bera(residuals)
print(f"Jarque-Bera: {jb_stat:.2f}, p-value: {jb_p:.3e}")

# Skew and Kurtosis
print(f"Skew: {residuals.skew():.3f}, Kurtosis: {residuals.kurtosis():.3f}")
```
