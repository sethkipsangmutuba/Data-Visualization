# Aggregations: Min, Max, and Everything in Between

Aggregation functions in NumPy reduce an array of values into a single summary statistic. These operations are fundamental in statistical analysis and data exploration, as they provide compact descriptions of large datasets.

Common aggregation functions include:
- Sum (total magnitude)
- Minimum (smallest value)
- Maximum (largest value)
- Mean (average)
- Standard deviation and variance (spread of values)

These operations are optimized and implemented in compiled code, making them significantly faster than equivalent Python loop-based implementations.

---

## Summing the Values in an Array

Summation computes the total of all elements in an array. In NumPy, this is performed using an optimized reduction operation that avoids explicit iteration.

The summation operation works across:
- Entire arrays (global sum)
- Specific axes in multidimensional arrays (row-wise or column-wise sums)

This flexibility makes it a core tool in statistical computation, feature engineering, and numerical modeling.

---

## Minimum and Maximum

Minimum and maximum operations identify the smallest and largest values in an array, respectively.

These functions are widely used for:
- Range estimation
- Data normalization
- Outlier detection
- Preprocessing in machine learning pipelines

Like other NumPy aggregations, these operations are vectorized and computationally efficient, enabling rapid analysis of large datasets.

##4.1) Summing the Values in an Array

NumPy provides `np.sum`, which performs the same operation but much faster due to optimized compiled code.

For large arrays, NumPy’s approach is significantly more efficient.

The two functions are not identical: they differ in behavior for multi-dimensional arrays and optional parameters.

## Minimum and Maximum

Python provides basic functions to find minimum and maximum values in arrays.

NumPy offers equivalent functions that are much faster on large datasets.

NumPy versions are optimized and suitable for large-scale numerical computation.

Arrays also support direct method-style calls for `min`, `max`, and `sum`.

Important practice: always prefer NumPy aggregation functions when working with NumPy arrays for performance and consistency.

```python
import numpy as np

# -----------------------------
# Iris dataset
# -----------------------------
x = iris_df.iloc[:, 0].to_numpy()
X = iris_df.iloc[:, :4].to_numpy()

# 1. Python built-in min/max (scalar iteration model)

python_min = min(x)
python_max = max(x)
```

```python
# 2. NumPy min/max (vectorized reduction kernels)

numpy_min = np.min(x)
numpy_max = np.max(x)

print("Python min:", python_min)
print("Python max:", python_max)

print("NumPy min:", numpy_min)
print("NumPy max:", numpy_max)
```

```python
# 3. Method-style calls (NumPy array methods)

print("\nMethod-style min:", x.min())
print("Method-style max:", x.max())
print("Method-style sum:", x.sum())
```

```python
# 4. Multi-dimensional behavior (axis-aware reductions)

col_min = np.min(X, axis=0)
col_max = np.max(X, axis=0)

row_min = np.min(X, axis=1)
row_max = np.max(X, axis=1)

print("\nColumn-wise min:", col_min)
print("Column-wise max:", col_max)

print("\nRow-wise min (first 5):", row_min[:5])
print("Row-wise max (first 5):", row_max[:5])
```

## Multidimensional Aggregates

Aggregation functions in NumPy can operate across specific dimensions (rows or columns) of a 2D array.

By default, aggregation is performed over the entire array, producing a single result.

The `axis` parameter controls the direction of aggregation:

- `axis=0` → operates down columns (collapses rows)  
- `axis=1` → operates across rows (collapses columns)  

Column-wise aggregation returns one value per column.

Row-wise aggregation returns one value per row.

Key idea: the chosen axis is the one that gets reduced (collapsed), not the one returned.

```python
import numpy as np

# Iris dataset (feature matrix)

X = iris_df.iloc[:, :4].to_numpy()

# 1. Global aggregation (no axis)

global_mean = np.mean(X)
global_sum = np.sum(X)

print("Global mean:", global_mean)
print("Global sum:", global_sum)
```

```python
# 2. Column-wise aggregation (axis=0)

col_mean = np.mean(X, axis=0)
col_min = np.min(X, axis=0)
col_max = np.max(X, axis=0)

print("\nColumn-wise mean:", col_mean)
print("Column-wise min:", col_min)
print("Column-wise max:", col_max)
```

```python
# 3. Row-wise aggregation (axis=1)

row_mean = np.mean(X, axis=1)
row_min = np.min(X, axis=1)
row_max = np.max(X, axis=1)

print("\nRow-wise mean (first 5):", row_mean[:5])
print("Row-wise min (first 5):", row_min[:5])
print("Row-wise max (first 5):", row_max[:5])
```

##4.2) Other Aggregation Functions (Brief Notes)

NumPy offers a wide range of aggregation functions beyond sum, min, and max.

Many functions have NaN-safe versions, which ignore missing values (NaN) during computation.

NaN-safe functions were introduced in newer NumPy versions (≥1.8).

---

### Key aggregation functions:

- `sum` / `nansum` → total of elements  
- `prod` / `nanprod` → product of elements  
- `mean` / `nanmean` → average value  
- `std` / `nanstd` → standard deviation  
- `var` / `nanvar` → variance  
- `min` / `nanmin` → minimum value  
- `max` / `nanmax` → maximum value  
- `argmin` / `nanargmin` → index of minimum  
- `argmax` / `nanargmax` → index of maximum  
- `median` / `nanmedian` → median value  
- `percentile` / `nanpercentile` → distribution-based statistics  
- `any` → checks if any element is True  
- `all` → checks if all elements is True  

---

### Core insight:

Use NaN-safe functions when working with incomplete or messy datasets.

```python
import numpy as np

# -----------------------------
# Iris dataset
# -----------------------------
X = iris_df.iloc[:, :4].to_numpy()

# Introduce NaNs artificially (for demonstration)
X_nan = X.copy()
X_nan[0, 0] = np.nan
X_nan[10, 1] = np.nan
X_nan[20, 2] = np.nan

# 1. Sum vs NaN-safe sum

print("Normal sum (may propagate NaN):", np.sum(X_nan))
print("NaN-safe sum:", np.nansum(X_nan))
```

```python
# 2. Product vs NaN-safe product

print("\nNaN-safe product:", np.nanprod(X_nan[:5]))
```

```python
# 3. Mean, std, variance

print("\nMean:", np.mean(X_nan))
print("NaN-safe mean:", np.nanmean(X_nan))

print("Std:", np.std(X_nan))
print("NaN-safe std:", np.nanstd(X_nan))

print("Variance:", np.var(X_nan))
print("NaN-safe variance:", np.nanvar(X_nan))
```

```python
# 4. Min / Max and index tracking

print("\nArgmin index:", np.nanargmin(X_nan))
print("Argmax index:", np.nanargmax(X_nan))

print("Min value:", np.nanmin(X_nan))
print("Max value:", np.nanmax(X_nan))
```

```python
# 5. Median and percentile

print("\nMedian:", np.nanmedian(X_nan))
print("75th percentile:", np.nanpercentile(X_nan, 75))
```

```python
# 6. Boolean aggregation (any / all)

print("\nAny > 7:", np.any(X_nan > 7))
print("All > 0:", np.all(X_nan > 0))
```
