## Fancy Indexing

In previous sections, we accessed parts of arrays using simple indexing (single values), slicing (ranges of values), and Boolean masking (filter-based selection). Fancy indexing extends this idea by using arrays of indices instead of single values. This allows direct selection of multiple, non-contiguous elements and enables more flexible and powerful extraction or modification of array subsets.

##7.1) Exploring Fancy Indexing

Fancy indexing uses arrays of indices to access multiple elements at once.

Instead of single values or slices, you pass index lists or arrays to select elements.

The result depends on the shape of the index array, not the original array.

It works in multiple dimensions by pairing row and column index arrays element-wise.

With broadcasting, index arrays can combine to produce grid-like selections.

Key idea: fancy indexing enables flexible, non-contiguous selection of array elements efficiently.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()

# 1. 1D fancy indexing (arbitrary element selection)

indices = np.array([0, 5, 10, 20, 30])

selected_rows = X[indices]

print("Selected rows shape:", selected_rows.shape)
print("Selected rows:\n", selected_rows)
```

```python
# 2. Column selection via fancy indexing

col_indices = np.array([0, 2, 3])  # select specific features

selected_cols = X[:, col_indices]

print("\nSelected columns shape:", selected_cols.shape)
```

```python
# 3. Row + column fancy indexing (paired indexing)

row_idx = np.array([0, 1, 2, 3])
col_idx = np.array([3, 2, 1, 0])

paired_selection = X[row_idx, col_idx]

print("\nPaired indexing result:", paired_selection)
```

```python
# 4. Grid-style selection using broadcasting

row_grid = np.array([0, 1, 2])[:, np.newaxis]
col_grid = np.array([0, 1, 2])

grid_selection = X[row_grid, col_grid]

print("\nGrid selection shape:", grid_selection.shape)
```

```python
# 5. Non-contiguous random sampling

random_idx = np.random.RandomState(42).choice(150, size=10, replace=False)

random_sample = X[random_idx]

print("\nRandom sample shape:", random_sample.shape)
```

##7.2) Combined Indexing

Fancy indexing can be combined with simple indexing, slicing, and masking to create more flexible selections.

You can mix:

- Simple index + fancy index to reorder elements  
- Slicing + fancy index to select submatrices with custom columns  
- Boolean masks + fancy index to filter specific columns or rows  

These combinations allow highly flexible data extraction and modification.

---

## Example: Selecting Random Points

Fancy indexing is commonly used to select random subsets of data points from a dataset. For a matrix representing points in a 2D space, random indices are generated and used to extract a subset of rows.

This is useful for sampling and splitting data in statistical analysis and model evaluation.

---

## Key idea

Combining indexing methods enables powerful, flexible data selection strategies.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()

# 1. Simple index + fancy index (reordering rows)

simple_idx = 2
fancy_idx = np.array([0, 5, 10, 15])

reordered = X[[simple_idx, *fancy_idx]]

print("Reordered rows shape:", reordered.shape)
```

```python
# 2. Slicing + fancy indexing (custom column selection)

subset = X[:, 1:4]                 # slice first
fancy_cols = np.array([0, 2])      # within subset

combined_cols = subset[:, fancy_cols]

print("\nSlicing + fancy indexing shape:", combined_cols.shape)
```

```python
# 3. Boolean mask + fancy indexing (filtered + selected rows)

mask = X[:, 0] > 5.5              # filter condition
filtered = X[mask]                # step 1: masking

fancy_rows = np.array([0, 2, 4])  # step 2: fancy selection

final_selection = filtered[fancy_rows]

print("\nMask + fancy indexing shape:", final_selection.shape)
```

```python
# 4. Random sampling using fancy indexing

rng = np.random.RandomState(42)

random_indices = rng.choice(X.shape[0], size=10, replace=False)

random_sample = X[random_indices]

print("\nRandom sample shape:", random_sample.shape)
print("Random sample:\n", random_sample)
```

```python
# 5. Mixed row/column indexing (advanced selection)

row_idx = np.array([0, 1, 2])
col_slice = slice(1, 3)

mixed_selection = X[row_idx][:, col_slice]

print("\nMixed indexing shape:", mixed_selection.shape)
```

## Modifying Values with Fancy Indexing

Fancy indexing can also be used to change array values directly by assigning to selected indices.

You can:

- Assign a single value to multiple indexed positions  
- Use assignment operators (e.g., `+=`, `-=`) on selected indices  

However, repeated indices can produce unexpected behavior because assignments are applied in a single step, not sequentially.

To ensure repeated updates work correctly (i.e., applying an operation multiple times per index), NumPy provides ufunc methods like `at()`, which perform in-place updates per occurrence.

---

## Example: Binning Data

Fancy indexing can be used to build histograms by:

- Defining bin edges  
- Finding bin positions using search operations  
- Using `ufunc.at()` to increment counts efficiently  

This reproduces histogram-like behavior manually and helps illustrate how binning works internally.

In practice, built-in functions like `histogram` are preferred because they are optimized and more flexible.

---

## Key idea

Fancy indexing allows both direct modification of selected elements and low-level control over aggregation-like operations.

```python
import numpy as np

# Iris dataset

X = iris_df.iloc[:, :4].to_numpy()

# Work on a copy to preserve original
x = X[:, 0].copy()

# 1. Assign single value to multiple indices

idx = np.array([0, 5, 10, 15])

x[idx] = 100

print("After assignment (first 20):", x[:20])
```

```python
# 2. Augmented assignment (+=)

x2 = X[:, 0].copy()

x2[idx] += 10

print("\nAfter += (first 20):", x2[:20])
```

```python
# 3. Repeated indices problem

x3 = X[:, 0].copy()

repeated_idx = np.array([0, 0, 0, 1, 1])

x3[repeated_idx] += 1

print("\nRepeated index result:", x3[:5])
```

Binning

```python
data = X[:, 0]  # sepal length

# Define bins
bins = np.linspace(data.min(), data.max(), 6)

# Find bin indices
bin_idx = np.searchsorted(bins, data) - 1

# Initialize counts
counts = np.zeros(len(bins) - 1, dtype=int)

# Accumulate counts
np.add.at(counts, bin_idx, 1)

print("Manual histogram counts:", counts)
print("Bins:", bins)
```
